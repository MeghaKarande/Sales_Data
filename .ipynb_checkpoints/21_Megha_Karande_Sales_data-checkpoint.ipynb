# Model Building

```python
df1
```

```python
x = df1.drop("Item_Outlet_Sales", axis = 1)
y = df1["Item_Outlet_Sales"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 15)
```

```python
model = LinearRegression()     # creating instance
model
```

```python
model.fit(x_train, y_train)     # model fitting
```

# Model Evaluation

```python
y_pred_train = model.predict(x_train)                        # training data

mse = mean_squared_error(y_train, y_pred_train)
print("MSE :", mse)

rmse = np.sqrt(mse)
print("RMSE :", rmse)

mae = mean_absolute_error(y_train, y_pred_train)
print("MAE :", mae)

r2 = r2_score(y_train, y_pred_train)
print("R2 score:", r2)
```

```python
y_pred_test = model.predict(x_test)                      # testing data

mse = mean_squared_error(y_test, y_pred_test)
print("MSE :", mse)

rmse = np.sqrt(mse)
print("RMSE :", rmse)

mae = mean_absolute_error(y_test, y_pred_test)
print("MAE :", mae)

r2 = r2_score(y_test, y_pred_test)
print("R2 score:", r2)
```

## 3. Normality of Residual

```python
Residual = y_train - y_pred_train
Residual
```

```python
sns.kdeplot(Residual)                                               # kdeplot
```

```python
from scipy.stats import shapiro                                  # shapiro test
```

```python
_, p_val = shapiro(Residual)

if p_val >= 0.05:
    print("Null Hypothesis is Accepted")
    print("Data is Normally Distributed")
    
else:
    print("Alternate Hypothesis is Accepted")
    print("Data is NOT Normally Distributed")
```

```python
import statsmodels.api as sm
```

```python
sm.qqplot(Residual, line = "45", fit = True)
```
